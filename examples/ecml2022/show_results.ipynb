# Show results

Tables show errors for particular number of training examples (rows) and data split (column). Two right most columns show mean and std over the splits.

```python
#results_root = "/home/xfrancv/Work/ConsistentSurrogate/code/PyMaxSum/results/"
results_root = "results/"

results = []
for e in exp:
    folder = f"{results_root}{e['dataset']}/{e['setup']}/"
    results.append( ut.get_results( folder, lambda_range=(0,0) ) )
    
       
plt.figure()
print("[testing error - hamming]")
for i, e in enumerate( exp ):
    results[i]["tst_err"]["mean"].plot(logx=True,label=e["label"])
    print( e["label"] )
    display( results[i]["tst_err"])
    print("")
plt.grid('on')
plt.title("test error - hamming")
plt.ylabel("test error [hamming]")
plt.xlabel('# training examples')
plt.legend()
    
plt.figure()
print("[training error - hamming]")
for i, e in enumerate( exp ):
    results[i]["trn_err"]["mean"].plot(logx=True,label=e["label"])
    print( e["label"])
    display( results[i]["trn_err"])
    print("")
plt.grid('on')
plt.title("training error - hamming")
plt.ylabel("training error [hamming]")
plt.xlabel('# training examples')
plt.legend()

plt.figure()
print("[testing error - 01]")
for i, e in enumerate( exp ):
    results[i]["tst_err_01"]["mean"].plot(logx=True,label=e["label"])
    print( e["label"])
    display( results[i]["tst_err_01"])
    print("")
plt.grid('on')
plt.title("test error - 01")
plt.ylabel("test error [0/1]")
plt.xlabel('# training examples')
plt.legend()

plt.figure()
print("[training error - 01]")
for i, e in enumerate( exp ):
    results[i]["trn_err_01"]["mean"].plot(logx=True,label=e["label"])
    print( e["label"])
    display( results[i]["trn_err_01"])
    print("")
plt.grid('on')
plt.title("training error - 01")
plt.ylabel("training error [0/1]")
plt.xlabel('# training examples')
plt.legend()


plt.figure()
for i, e in enumerate( exp ):
    results[i]["best_lambda"]["mean"].plot(logx=True,label=e["label"])
plt.grid('on')
plt.title("best lambda")
plt.legend()

```
